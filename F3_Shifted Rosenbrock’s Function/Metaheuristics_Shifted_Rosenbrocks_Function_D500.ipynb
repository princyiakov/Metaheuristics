```python
def func(x):
    F = 0
    y = np.empty(D)
    for i in range(D - 1):
        y[i] = x[i] - data[i] +1
    for i in range(D-2):
        F += 100 * ((y[i]**2-y[i+1])**2) + (y[i]-1)**2
    func = F + f_bias
    return func
```

```python
class Rosenbrocks_problem:
    def __init__(self,xmin,xmax,D,f_bias):
        self.lower_bound = xmin
        self.upper_bound = xmax
        self.dimension = D
        self.glob_opt = f_bias

    def fitness(self, x):
        result = func(x)
        return [result]

    def get_bounds(self):
        x_min = self.lower_bound * np.ones(self.dimension)
        x_max = self.upper_bound * np.ones(self.dimension)
        return x_min, x_max
```

```python
def Solution_de(func):
    gen=20000
    start_time = time.time()
    algo = pg.algorithm(pg.de(gen=gen, 
                          F=0.6, 
                          CR=0.95, 
                          variant=3, 
                          ftol=1e-5, 
                          xtol=1e-5
                         ))
    algo.set_verbosity(5)
    popsize = 5
    prob = pg.problem(Rosenbrocks_problem(xmin,xmax,D,f_bias))
    pop = pg.population(prob, popsize)
    extract_algo = algo.extract(pg.de)
    pop = algo.evolve(pop)
    end_time = time.time()
    
    log = extract_algo.get_log()
    curve = [x[2] for x in log]
    niter = pop.problem.get_fevals()
    plt.plot(curve)
    
    plt.title("Shifted Rosenbrocks Function", fontsize=20,fontweight='bold')
    plt.xlabel("Iteration", fontsize=18,fontweight='bold')
    plt.ylabel("Function", fontsize=18,fontweight='bold')
    plt.show()
    print(pop)
    print("Computational time: {:.2f} seconds".format(time.time()-start_time))
    print("Best position: ", pop.champion_x)
    print("Best Fitness",pop.champion_f[0])
    print("Number of iterations ",niter)
```
