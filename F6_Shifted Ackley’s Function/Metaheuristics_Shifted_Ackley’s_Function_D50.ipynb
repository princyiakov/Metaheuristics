```python
def func(x):
    Sum1 = 0
    Sum2 = 0
    for i in range(D - 1):
        z = x[i] - data[i]
        Sum1 += z ** 2
        Sum2 += math.cos(2 * math.pi * z)
    func = -20 * math.exp(-0.2 * math.sqrt(Sum1 / D)) - math.exp(Sum2 / D) + 20 + math.e + f_bias
    return func
```

```python
class Ackley_problem:
    def __init__(self,xmin,xmax,D,f_bias):
        self.lower_bound = xmin
        self.upper_bound = xmax
        self.dimension = D
        self.fitness_per_eval = []
        self.glob_opt = f_bias

    def fitness(self, x):
        #result = abs(func(x)-f_bias)
        result = func(x)
        return [result]

    def get_bounds(self):
        x_min = self.lower_bound * np.ones(self.dimension)
        x_max = self.upper_bound * np.ones(self.dimension)
        return x_min, x_max
```

```python
def Solution_de(func):
    gen=150
    start_time = time.time()
    algo = pg.algorithm(pg.de(gen=gen, 
                          F=0.6, 
                          CR=0.95, 
                          variant=3, 
                          ftol=0, 
                          xtol=1e-5
                         ))
    algo.set_verbosity(100)
    popsize = 100
    prob = pg.problem(Ackley_problem(xmin,xmax,D,f_bias))
    pop = pg.population(prob, popsize)
    extract_algo = algo.extract(pg.de)
    pop = algo.evolve(pop)
    end_time = time.time()
    
    log = extract_algo.get_log()
    niter = pop.problem.get_fevals()
    
    curve = [x[2] for x in log]
    plt.plot(curve)
    #plt.plot([x[0] for x in log], [x[1] for x in log])
    plt.title("Shifted Ackley’s Function", fontsize=20,fontweight='bold')
    plt.xlabel("Iteration", fontsize=18,fontweight='bold')
    plt.ylabel("Cost", fontsize=18,fontweight='bold')
    plt.show()
    print(pop)
    print("Computational time: {:.2f} seconds".format(time.time()-start_time))
    print("Best position: ", pop.champion_x)
    print("Best Fitness",pop.champion_f[0])
    print("Number of iterations ",niter)
```
